```python
import threading

group = [-2, -3, 15, 14, 7, -10, 20, 4, -2, 76, -89, 102, 221, -333, 1, -5, 67]
target = 0

def permutations(group):
    LENGTH = len(group)
    groups = []
    iters = 2**LENGTH

    for i in range(1, iters):
        s = str(bin(i))[2:]
        for i in range(LENGTH - len(s)):
            s = '0' + s

        pos = [i for i, v in enumerate(s) if v == '1']
        groups.append([group[i] for i in pos])
    
    return groups


def agent(subgroup, var):
    s = sum(subgroup)
    var.append({'s': s, 'g': subgroup})


threads = []
sums = []

perms = permutations(group)

for perm in perms:
    t = threading.Thread(target=agent, args=(perm, sums,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

results = []
for i in sums:
    if i['s'] == target:
        results.append(i)

for i in results:
    print('{} <= {}'.format(i['s'], i['g']))
```
